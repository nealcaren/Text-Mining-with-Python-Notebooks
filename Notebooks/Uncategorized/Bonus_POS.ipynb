```python
import pandas as pd

wine_df_full = pd.read_csv('data/wine_reviews.csv')
```

```python
wine_df = wine_df_full.sample(10000).reset_index()
```

```python
wine_df['doc'] = wine_df['description'].apply(nlp)
```

```python
from pdtext.st import nlp_col_2_df
```

```python
long = nlp_col_2_df(wine_df['doc'])
```

```python
long.head(20)
```

```python
long['upos'].value_counts()
```

```python
long[long['upos'] == 'ADJ']['lemma'].value_counts().head()
```

```python
long[long['upos'] == 'VERB']['lemma'].value_counts().head()
```

```python
def just_adjectives(doc):
    adjs = []
    for sentence in doc.sentences:
        for word in sentence.words:
            if word.upos == 'ADJ':
                adjs.append(word.text)
    return ' '.join(adjs)
    return 
            
```

```python
wine_df['doc'].apply(just_adjectives)
```

```python
wine_df['adjs'] = wine_df['doc'].apply(just_adjectives)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression

v = CountVectorizer(min_df=10)
v.fit(wine_df['adjs'])
atf = v.transform(wine_df['adjs'])

lr = LinearRegression()
lr.fit(atf, wine_df['points'])

coeficients = pd.Series(lr.coef_,
                        index = v.get_feature_names())

coeficients.sort_values(ascending=False)[:5]
```

```python
coeficients.sort_values(ascending=True)[:5]
```
