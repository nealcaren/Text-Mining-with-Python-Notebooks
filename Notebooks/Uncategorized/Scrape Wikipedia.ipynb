```python
from requests_html import HTMLSession

import pandas

def get_text(parsed_html):
    article_text = ''
    for paragraph in parsed_html.find('p'):
        new_paragraph = paragraph.text
        article_text  = article_text + '\n' + new_paragraph
    return article_text

def get_headline(parsed_html):
    headline = parsed_html.find('h1')[0].text
    return headline


def scrape_wikipedia(url):
    '''Scrape a Wikipedia page and return the headline and text.'''
    
    # gets the article
    session = HTMLSession()
    r = session.get(url)

    # parses the html
    parsed_html = r.html
    
    # grabs the headline
    headline = get_headline(parsed_html)

    # grabs the text
    text = get_text(parsed_html)
    
    # stores headline and text in a dictionary
    results = {'headline' : headline,
               'text'     : text,
               'url'      : url}
    
    # send it back
    return results
```

```python
scrape_wikipedia('https://en.wikipedia.org/wiki/Sociocultural_evolution')
```

### Scraping the Sun

```python
def get_text(parsed_html):
    article_text = ''
    for paragraph in parsed_html.find('p')[1:-13]:
        new_paragraph = paragraph.text
        article_text  = article_text + '\n' + new_paragraph
    return article_text

def get_headline(parsed_html):
    headline = parsed_html.find('h1')[0].text
    return headline

def get_author(parsed_html):
    author = parsed_html.find('.article__author-name')[0].text
    return author

def scrape_thesun(url):
    '''Scrape a The Sun article and return the headline and text.'''
    
    # gets the article
    session = HTMLSession()
    r = session.get(url)

    # parses the html
    parsed_html = r.html
    
    # grabs the headline
    headline = get_headline(parsed_html)

    author = get_author(parsed_html)
    
    # grabs the text
    text = get_text(parsed_html)
    
    # stores headline and text in a dictionary
    results = {'headline' : headline,
               'text'     : text,
               'url'      : url,
              'author' : author}
    
    # send it back
    return results

```

```python
scrape_thesun(url)
```
